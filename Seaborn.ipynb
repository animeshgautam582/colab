# Seaborn

Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

Seaborn has some built in data set

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting theme
sns.set_theme()
```

```python
#visualizing tips dataframe
sns.relplot(data=tips, x ='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
```

```python
#loading iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.scatterplot(x='sepal_length', y='petal_width', hue='species', data=iris)
```

```python
#loading the titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

Count Plot

```python
sns.countplot(x='class', data=titanic)
```

```python
sns.countplot(x='survived', data=titanic)
```

```python
sns.countplot(x='sex', data=titanic)
```

Bar Chart

```python
sns.barplot(x='sex',y='survived',hue='class', data=titanic)
```

```python
#house price dataset
from sklearn.datasets import load_wine

wined = load_wine()
wine = pd.DataFrame(wined.data, columns=wined.feature_names)
wine.head()
```

Distribution Plot

```python
sns.displot(wine['proline'])
```

Correlation Matrix :
1.Positive
2.Negative

Using Heat Map Plot

```python
correlation = wine.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
